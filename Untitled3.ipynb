```python
import numpy as np
import pandas as pd
import os
```

```python
datapath = 'Data/ner_tagged_news/'
```

```python
datapaths = os.listdir(datapath)
```

```python
df = pd.DataFrame()
for l in datapaths:
    dataFrame = pd.read_csv(datapath + l)
    
    df = df.append(dataFrame)
```

```python
df
```

```python
df = df.reset_index(drop=True)
df['index'] = df.index
```

```python
df = df.drop('Unnamed: 0', axis=1)
```
