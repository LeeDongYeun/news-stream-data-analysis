```python
df = pd.read_csv(datapath)
```

```python
df[' time'] = pd.to_datetime(df[' time'])
df = df.sort_values([' time'])
df = df.set_index(' time')
```

```python
df['2017-09-17']
```

```python
l = list(df['2017-09-17']['Organization'])
```

```python
word_list = []
for x in l:
    if type(x) == str:
        word_list += x.split(',')
```

```python
word_list
```

```python
except_set = ['(yonhap)', 'yonhap', '']
word_list = [x for x in word_list if x not in except_set]
```

```python
string.punctuation
```

```python
punc = '!"#$%&\'()*+,-/:;<=>?@[\\]^_`{|}~'
punc += "“”’“"
word_list = [word.translate(word.maketrans(punc, " "*len(punc))).strip() for word in word_list]
```
